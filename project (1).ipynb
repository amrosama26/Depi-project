```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Sales_dataset (1).csv')
```

```python
df
```

# nulls= city,postal_code,sales

```python
df.isnull().sum()
```

# fill the null values of City by mode

```python
df['City']=df['City'].fillna(df['City'].mode()[0])
```

```python
df.isnull().sum()
```

# Convert the 'Sales' column to numeric



```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
```

# fill the null of sales by mean

```python
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
df.isnull().sum()
```

# dropping column postal code

```python
df.drop(columns=['Postal Code'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df
```

# convert data type of orderdate to date time

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

# convert data type of ship date to date time

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'],dayfirst=True,errors='coerce')
```

# making a new column "day" of order date

```python
df['Day_of_order']=df['Order Date'].dt.day
```

# making a new column "month" of order date


```python
df["Month_of_order"]=df['Order Date'].dt.month
```

# making a new column "Name of day" of order date

```python
df['Day Name']=df['Order Date'].dt.day_name()
```

# making a new column "month" of order date

```python
df["Month Name"]=df['Order Date'].dt.month_name()
```

# making a new column "year" of order date


```python
df["Year"]=df['Order Date'].dt.year
```

# making a new column "deliery delays" of order date


```python
df["Delivery Delay (Dayes)"] = (df["Ship Date"] - df["Order Date"])
```
